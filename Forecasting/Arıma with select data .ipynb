```python
def clean_limit(df_column, low_limit, up_limit=None):
    if up_limit is None:
        filtered = df_column[(df_column >= low_limit)]
        
    else:
        filtered = df_column.loc[(df_column >= low_limit) & (df_column <= up_limit)]
    return filtered
    
```

## 3 sigma

```python
process_clm = slct_df["xVRMS"]
process_clm_mean = process_clm.mean()
process_clm_std = process_clm.std()

up_limit = process_clm_mean + 3 * process_clm_std
low_limit = process_clm_mean - 3 * process_clm_std

process_clm_mean,process_clm_std,up_limit,low_limit
```

## Use to Arıma

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(slct_df, model='multiplicative',period=2016)
result.plot()
plt.show()
```

```python
time = pd.date_range("2022-09-10 09:00:00", periods=len(slct_df), freq="5min")
values = np.array(slct_df.values)
series = pd.Series(values, index=time)
print(series)

# Günlük ortalamaya dönüştürme
daily_mean = series.resample("D").mean()
```

```python
len(daily_mean)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
# Define the model
model = SARIMAX(daily_mean, order=(2,1,3), seasonal_order=(0, 0, 0, 12))

# Fit the model
results = model.fit()

# Make predictions
pred = results.get_prediction(start="2022-10-27" , end='2022-11-05 ')
pred_ci = pred.conf_int()

# Plot the results
ax = daily_mean.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='forecast', alpha=.7)
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('xVRMS')
plt.legend()
plt.show()
```
